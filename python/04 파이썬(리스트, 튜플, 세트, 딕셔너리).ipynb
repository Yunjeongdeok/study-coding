## 리스트(List)



```
print([])
print([1,2,3])
print(['One', 'Two', 'Three'])
print([1,2,'One',3,4, 'Two'])
print([1,2,3, ['one','two',2,4]])
```

### 리스트 인덱싱(List Indexing)

* 리스트의 각 위치에 해당하는 값에 접근하기 위해서 주소 개념의 숫자를 사용

```
list = ['One', 'Two', 'Three','four']
print(list)
print(list[0])
print(list[1])
print(list[2])
print(list[3])
print(list[-1])
print(list[-2])
print(list[-3])
```

### 중첩 리스트 인덱싱(Nested List Indexing)

* 리스트 안에 리스트가 있을 경우 인덱스 접근 방법

```
list =[1,2,3, ['one','two',2,4]]
print(list)
print(list[3])
print(list[3][0])
print(list[3][1])
print(list[3][2])
print(list[3][3])
```

### 리스트 슬라이싱(List Slicing)

* 리스트의 인덱스를 통해서 부분만 가져올 때 사용

```
list = ['One', 'Two','Three']
print(list)
print(list[0:])
print(list[1:2])
print(list[1:])
```

### 중첩 리스트 슬라이싱(Nested List Slicing)

* 중첩된 리스트에서 일부분만 인덱스를 통해서 가져올 때 사용

```
list = [1,2,3, ['one','two',2,4]]
print(list)
print(list[2:4])
print(list[3][2:])
print(list[3][:2])
```

### 리스트 연산자/함수(List Operators/Function)

* 더하기 + 연산자
* 곱하기 * 연산자
* 리스트 길이를 구하는 `len()` 함수

```
list_1 = ['One', 'Two','Three']
list_2 = ['Four', 'Five', 'Six']

print(list_1)
print(list_2)
print(list_1 + list_2)

print(list_1 * 3)

print(len(list_1))
print(len(list_1 * 3))
```

### 리스트 수정(List Modify)

* 리스트 인덱스 주소를 이용한 값 수정

```
list = ['One', 'Two','Three']
print(list)
list[2] = 3
print(list)
list[1] = 2
print(list)
list[0] = 1
print(list)
```

### 리스트 메소드(List Methods)


#### append()

* 리스트 요소 추가

```
list = ['One', 'Two','Three']
list.append('Four')
print(list)

list.append([1,2,3,4])
print(list)
```

#### sort()

* 리스트 정렬

```
list = [13,34,243,1,5,64,127,234,6,12343,8,9,754,212]
print(list)
list.sort()
print(list)

list = ['orange', 'apple', 'Banana', 'Apple', 'strawberry']
list.sort()
print(list)
```

#### reverse()

* 리스트 요소 반전

```
list = ['orange', 'apple', 'Banana', 'Apple', 'strawberry']
print(list)
list.reverse()
print(list)
```

#### index()

* 리스트의 요소 값에 대한 인덱스를 반환

```
list = [9,8,7,6,5,4,3,2,1,0]
print(list)
print(list.index(7))
print(list.index(9))
```

#### insert()

* 리스트 요소 삽입

```
list = [9,8,7,6,5,4,3,2,1,0]
print(list)
list.insert(1,100)
print(list)
list.insert(0,99)
print(list)
```

#### remove()

* 리스트 요소 제거

```
list = [8,9,8,7,6,5,4,3,2,1,0]
print(list)
list.remove(8)
print(list)
list.remove(7)
print(list)
```

#### del

* 리스트 요소 제거 연산

```
list = [8,9,8,7,6,5,4,3,2,1,0]
print(list)
del list[0]
print(list)
del list[1]
print(list)
```

#### pop()

* 리스트 요소를 방출

```
list = [8,9,8,7,6,5,4,3,2,1,0]
print(list)
list.pop()
print(list)
list.pop(0)
print(list)
```

#### count()

* 리스트 요소의 갯수 계산

```
list = [10, 20, 20, 30, 30, 30]
print(list)
print(list.count(20))
print(list.count(30))
```

#### extend()

* 리스트 확장

```
list = [10, 20, 30, 40]
print(list)
list.extend([50,60])
print(list)
```

## 튜플(Tuple)

* 리스트와 유사하지만 변경 불가능한 시퀀스 자료형
* 하나의 변수에 여러 값 할당 가능
* ‘(‘와 ‘)’를 사용하여 표현

```
print(())
print((1,2,3))
print(('one', 'two', 'three'))
print((1, 'one', 2,'two',3, 'three'))
print((1,2,3,('one', 'two', 'three')))
```

### 튜플 인덱싱(Tuple Indexing)

* 튜플의 각 위치에 해당하는 값에 접근하기 위해서 주소 개념의 숫자를 사용

```
tuple = ('one', 'two', 'three')
print(tuple)
print(tuple[0])
print(tuple[-1])
```

### 중첩 튜플 인덱싱(Nested Tuple Indexing)

* 튜플 안에 튜플이 중첩되어 있을 경우, 인덱스 접근 방법

```
tuple = (1,2,3, ('one', 'two', 'three'))
print(tuple)
print(tuple[3])
print(tuple[3][1])
```

### 튜플 슬라이싱(Tuple Slicing)

* 튜플의 인덱스를 통해서 부분만 가져올 때 사용

```
tuple = ('one', 'two', 'three')
print(tuple)
print(tuple[0])
print(tuple[0:1])
print(tuple[2:])
print(tuple[:])
```

### 중첩 튜플 슬라이싱(Nested Tuple Slicing)

* 중첩된 튜플에서 일부분만 인덱스를 통해서 가져올 때 사용

```
tuple = (1,2,3, ('one', 'two', 'three'))
print(tuple)
print(tuple[3])
print(tuple[3][2:])
print(tuple[3][:2])
```

### 튜플 연산자/함수(Tuple Operators/Function)

* 더하기 + 연산
* 곱하기 * 연산
* 튜플 길이를 구하는 `len()` 함수



```
tuple_1 = ('one', 'two', 'three')
tuple_2 = ('four', 'five', 'six')

print(tuple_1)
print(tuple_2)
print(tuple_1 + tuple_2)
print(tuple_1 * 3)
print(len(tuple_1))
print(len(tuple_1 + tuple_2))
```

## 세트(Set)

* 데이터 중복을 허용하지 않는 구조
* 순서가 없는 데이터 집합을 위한 구조
* 인덱싱으로 값을 접근할 수 없음

```
print({})
print({'one', 'two', 'three'})
print({1,2,3,4,5})
print({3,2,4,1,2,3,3,4,5,5,6})
```

### 세트 연산자(Set Operators)

* 교집합: &
* 합집합: |
* 차집합: -
* 여집합: ^

```
set_1 ={1,2,3,4,5,6,6,7,8,9}
set_2 ={3,4,5,6,7,7,8,8,9,9}
print(set_1)
print(set_2)
print('교집합:', set_1 & set_2)
print('합집합:', set_1 | set_2)
print('차집합:', set_1 - set_2)
print('여집합:', set_1 ^ set_2)
```

### 세트 메소드(Set Methods)

* 교집합: intersection()
* 합집합: union()
* 차집합: difference()
* 여집합: symmetric_difference()

```
set_1 ={1,2,3,4,5,6,6,7,8,9}
set_2 ={3,4,5,6,7,7,8,8,9,9}
print(set_1)
print(set_2)

print(set_1.intersection(set_2))
print(set_1.union(set_2))
print(set_1.difference(set_2))
print(set_1.symmetric_difference(set_2))
```

* 요소 추가: add()
* 여러 요소 추가: update()
* 요소 제거: remove()
* 요소 제거: discard()
* 모든 요소 제거: clear()

```
set = {1,2,3,4}
print(set)

set.add(50)
print(set)

set.update([60, 70])
print(set)

set.remove(70)
set.remove(60)
print(set)

set.discard(1)
print(set)

set.clear()
print(set)
```

## 딕셔너리(Dictionary)

* 키(key)와 값(value)의 쌍으로 구성된 데이터
* 순서가 없는 데이터
* 키를 통해 값을 얻음
* 동일한 키가 있을경우 덮어씀

```
dic = {1:'One', 2:'Two', 3:'Three'}
print(dic)

dic = {1:'One, 1234', 2:'Two', 3:'Three',1:'One', 2:'Two', 3:'Three'}
print(dic)

dic = {1:'One, 1234', 2:'Two', 3:'Three',1:'One, 123', 2:'ABC', 3:'Three'}
print(dic)
```

### 딕셔너리 요소 추가/삭제

* 딕셔너리의 해당 키 값에 값을 추가하여 요소 추가
* del을 이용하여 요소 제거

```
dic = {1:'One', 2:'Two', 3:'Three'}
print(dic)
print(dic[2]) # [2]는 '키 값'이다.
dic[4] = 'Four'
print(dic)
dic[5] = 'Five', 5
print(dic)


del dic[1]
print(dic)
```

### 딕셔너리 메소드(Dictionary Methods)

|메소드|설명|
|---|---|
|keys()|키 가져오기
|values()|값 가져오기
|items()|키와 값 모두 가져오기
|get()|키에 해당하는 값 가져오기
|pop()|키에 해당하는 값 추출하기
|clear()|모든 요소를 제거하기



```
dic = {1:'One', 2:'Two', 3:'Three'}
print(dic)

print(dic.keys())
print(dic.values())
print(dic.items())
print(dic.get(2))
print(dic.pop(3))
print(dic)
dic.clear()
```
