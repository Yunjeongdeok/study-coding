## 매직 명령어(magic commands)

* Jupyter Notebook 환경에서 파이썬 파일 저장 및 실행

````
%%writefile test.py
print('hello colab')
````

````
%run test.py
````

* 코드 실행 시간 측정

````
%%timeit?
````

````
%timeit li = [n ** n for n in range(10000)]
````

````
%%timeit
 li = []
 for n in range(10000):
     li.append(n ** n)
````
