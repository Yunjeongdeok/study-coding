## 파이썬 시작하기

* 프로그래밍 언어를 배울 때 보통 가장 먼저 하는 것은 인사
* 가장 기본이 되는 화면 출력 함수인 `print()`를 이용해서 문자열 "Hello Python"을 출력

```
print("Hello Python")
```

## 파이썬으로 계산하기

### 사칙연산

* 파이썬에서는 기본적인 산술 연산을 바로 수행 가능
* 더하기, 빼기, 곱하기, 나누기 같은 간단한 산술연산 예제

```
1 + 1
```

```
5 - 1
```

```
4 * 3
```

```
3 / 5
```

### 실수 사칙연산

* 실수에 대해서도 사칙연산 가능

```
1.1 + 2.4
```

```
4.2 -1.2
```

```
1.2 - 4.2
```

```
5.4 - 2.4
```

### 복잡한 연산

* 여러 산술연산들이 포함된 복잡한 연산들도 가능

```
1 + 2 + 3 * 4 /3 *4
```

```
4 /2 * 2 -10 /4
```

```
(10-2 *4) * (3 - 6 - 3 + 3) / (3 * 5)
```

## 파이썬 문법(Python Syntax)


### 들여쓰기(Indentation)

* 코드 라인의 시작 부분에 들여쓰기
* 파이썬은 들여쓰기로 코드의 블록을 표시

```
if 2 > 1:
    print(2)
```

```
# 들여쓰기 안해서 에러가 났다.
# if 2 > 1:
# print(2)
```

### 주석(Comments)

* 코드에 대해 설명하는데 사용하고 가독성 증가
* 일부 코드를 주석처리하여 실행 방지 가능

```
# #은 주석을 뜻한다. 즉 코드라고 인식을 하지 않는다.
print('Yes!')
# print('No!') # 주석처리가 되어서 Yes!만 출력된다.
```
