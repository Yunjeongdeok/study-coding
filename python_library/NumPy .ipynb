### 배열 생성 및 초기화

- 문제, 아래의 함수 사용하기
    - zeros(), ones()
    - full(), eye(), tri(),empty() 

    - 아래의 함수 사용해보기
        - np.zeros_like()
        - np.ones_like()
        - np.full_like()
        - np.empty_like()

* `zeros()`: 모든 요소를 0으로 초기화

```
np.zeros(10)
np.zeros((3,4)) # np.array처럼 2,3차원은 겹을 쌓아야한다.
```

* `ones()`: 모든 요소를 1로 초기화

```
np.ones(5)
np.ones((4,4)) # np.array처럼 2,3차원은 겹을 쌓아야한다.
np.ones((4,5,6))
```

* `full()`: 모든 요소를 지정한 값으로 초기화

```
np.full((3,3), 1.23) # 단축키에 갖다대서 코랩의 설명을 잘 읽어보자.
```

* `eye()`: 단위행렬(identity matrix) 생성
  + 주대각선의 원소가 모두 1이고 나머지 원소는 모두 0인 정사각 행렬

```
print(np.eye(5))
print(np.eye(5).shape) # 2차원만 가능한건가? 모르겠다.
```

* `tri()`: 삼각행렬 생성

```
print(np.tri(3),np.tri(3).shape, sep='\n')
print('\n')
print(np.tri(5),np.tri(5).shape, sep='\n') # 2차원만 가능한건가? 모르겠다.
```

* `empty()`: 초기화되지 않은 배열 생성
  + 초기화가 없어서 배열 생성비용 저렴하고 빠름
  + 초기화되지 않아서 기존 메모리 위치에 존재하는 값이 있음

```
np.empty(10)
```

* `_like()`: 지정된 배열과 shape가 같은 행렬 생성
  + `np.zeros_like()`
  + `np.ones_like()`
  + `np.full_like()`
  + `np.empty_like()`

```
print(a1,'에서')
print(np.zeros_like(a1),'로 바뀌었다.') # a1의 행렬이 1차원이라서 1차원이다.
```

```
print(a2,'에서')
print(np.ones_like(a2),'로 바뀌었다.')# a2의 행렬이 2차원이라서 2차원이다
```

```
print(a3,'에서')
print(np.full_like(a3,10),'로 바뀌었다.')# a2의 행렬이 2차원이라서 2차원이다.
```

### 배열 속성 정보

```
def array_info(array):
    print(array)
    print("ndim:", array.ndim)
    print("shape:", array.shape)
    print("dtype:", array.dtype)
    print("size:", array.size)
    print("itemsize:", array.itemsize)
    print("nbytes:", array.nbytes)
    print("strides:", array.strides)
    
```

```
array_info(a1)
 # ndim인 이유는 [] 한 줄(1차원)이라서.
 # shape는 []에 나오는 그대로 보면 된다.
 # dtype도 그냥 확인해보니 int64.
 # size는 []안에 5개가 있으니까
 # itemsize는 8바이트라는 뜻이다. 왜냐하면 dtype이 int64이니까.
 # nbytes가 전체 바이트로 40인 이유는 size 5 * intemsize 8 =40이기 때문이다.
 # strides 설명을 들으니 머리로는 알겠는데 설명을 못하겠다.
```

## 배열 값 삽입/수정/삭제/복사

### 배열 값 삽입

* `insert()`: 배열의 특정 위치에 값 삽입
* axis를 지정하지 않으면 1차원 배열로 변환
* 추가할 방향을 axis로 지정
* 원본 배열 변경없이 새로운 배열 반환

```
print(a1)

b1 = np.insert(a1, 0, 10)
c1 = np.insert(a1, 2, 10)

print(a1)
print(b1) # insert로 인한 원본 훼손 X
print(c1) # insert로 인한 원본 훼손 X
```

```
print('a2', a2, sep='\n') ;print('\n')
b2 = np.insert(a2, 1, 10, axis=0)
c2 = np.insert(a2, 1, 10, axis=1)

print('b2', b2, sep='\n') ;print('\n')
print('c2', c2, sep='\n') ;print('\n')
```

```
print('a1', a1, sep='\n') ;print('\n')
a1[0] = 1
a1[1] = 4
a1[2] = 7
print('a1', a1, sep='\n') ;print('\n')

a1[:1] = 9
print('a1', a1, sep='\n') ;print('\n')

i = np.array([1, 3, 4])
a1[i] = 0
print('a1', a1, sep='\n') ;print('\n')

a1[i] += 4
print('a1', a1, sep='\n') ;print('\n')
```

```
print('a2', a2, sep='\n') ;print('\n')

a2[0,0] = 1
a2[1,1] = 2
a2[2,2] = 3
a2[0] = 1
print('a2', a2, sep='\n') ;print('\n')

a2[1:, 2] = 9
row = np.array([0, 1])
col = np.array([1, 2])
a2[row, col] = 0
print('a2', a2, sep='\n') ;print('\n')
```

### 배열 값 삭제

* `delete()`: 배열의 특정 위치에 값 삭제
* axis를 지정하지 않으면 1차원 배열로 변환
* 삭제할 방향을 axis로 지정
* 원본 배열 변경없이 새로운 배열 반환

```
print(a1) ;print('\n')
b1 = np.delete(a1,1)
print(b1);print('\n')
print(a1)
```

```
print(a2);print('\n')

b2 = np.delete(a2, 1, axis = 0)
c2 = np.delete(a2, 1, axis = 1)

print(b2);print('\n')
print(c2);print('\n')
```

### 배열 복사

* **리스트 자료형과 달리 배열의 슬라이스는 복사본이 아님 =>원본에도 영향을 끼친다.**

```
print(a2);print('\n')
print(a2[:2, :2]);print('\n')

a2_sub = a2[:2, :2]
print(a2_sub);print('\n')

a2_sub[:, 1] = 0 # 리스트와 달리 배열의 슬라이스는 원본에도 영향을 끼친다
print(a2_sub);print('\n') # 아래의 copy를 써야 원본에도 이상이 없다

print(a2);print('\n')
```


* `copy()`: 배열이나 하위 배열 내의 값을 명시적으로 복사

```
print(a2);print('\n')

a2_sub_copy = a2[:2, :2].copy()
print(a2_sub_copy);print('\n')

a2_sub_copy[:,1] = 1
print(a2_sub_copy);print(a2_sub_copy);print('\n')

print(a2);print('\n')
```

```
a1 = np.arange(1, 10)
print(a1)
print(a1 == 5) # 넘파이는 비교연산을 하면 불연산을 한다.
print(a1 != 5)
print(a1 > 5)
print(a1 >= 5)
print(a1 < 5)
print(a1 <= 5)
```

```
a2 = np.arange(1,10).reshape(3,3)
print('a2', a2, sep ='\n');print('\n')
print('<sum>')
print(np.sum(a2));print('\n')
print(np.count_nonzero(a2 > 5));print('\n')
print(np.sum(a2 > 5));print('\n')
print(np.sum(a2 > 5, axis = 0));print('\n')
print(np.sum(a2 > 5, axis = 1));print('\n')

print('<any>')
print(np.any(a2 > 5));print('\n')
print(np.any(a2 > 5, axis = 0));print('\n')
print(np.any(a2 > 5, axis = 1));print('\n')

print('<all>')
print(np.all(a2 > 5));print('\n')
print(np.all(a2 > 5, axis = 0));print('\n')
print(np.all(a2 > 5, axis = 1));print('\n')
```


#### 불리언 연산자(Boolean Operators)


|연산자|비교 범용 함수|
|---|---|
|&|np.bitwise_and|
|ㅣ|np.bitwise_or|
|^|np.bitwise_xor|
|~|np.bitwise_not|

```
a2 = np.arange(1,10).reshape(3,3)
print(a2)
print('\n', '<and>')
print((a2 > 5) & (a2 < 8))
print(a2[(a2 > 5) & (a2 < 8)])
print('\n', '<or>')
print((a2 > 5) | (a2 < 8))
print(a2[(a2 > 5) | (a2 < 8)])
print('\n', '<xor>')
print((a2 > 5) ^ (a2 < 8))
print(a2[(a2 > 5) ^ (a2 < 8)])
print('\n', '<not>')
print(~(a2 > 5))
print(a2[~(a2 > 5)])
```

### 배열 정렬

```
a1 = np.random.randint(1, 10, size = 10)

print('a1', a1, sep='\n')
print('\n', 'a1 정렬')
print(np.sort(a1))# 원본(a1) 유지
print('\n')
print('a1', a1, sep='\n')
print('\n', 'a1 arg정렬')
print(np.argsort(a1)) # 원본(a1) 유지
print('\n', 'a1')
print(a1)

print('\n', 'a1')
print(a1.sort()) # 원본(a1) 담음
print('\n', 'a1')
print(a1) # 원본(a1)이 정렬되었다.
```
